```python
def experiment():


    beta = 1

    precision_0 = []
    precision_1 = []
    recall_0 = []
    recall_1 = []
    accuracy = []
    error_rate = []
    false_positive_rate = []
    F_score = []
    type1 = []
    type2 = []

    clf = BernoulliNB()
    clf = clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    y_metrics = metrics.confusion_matrix(y_test, y_pred)
    print(y_metrics)
    precision_0_ = round(float(y_metrics[0, 0]) / float(y_metrics[0, 0] + y_metrics[1, 0]), 2)
    precision_1_ = round(float(y_metrics[1, 1]) / float(y_metrics[0, 1] + y_metrics[1, 1]), 2)
    recall_0_ = round(float(y_metrics[0, 0]) / float(y_metrics[0, 0] + y_metrics[0, 1]), 2)
    recall_1_ = round(float(y_metrics[1, 1]) / float(y_metrics[1, 0] + y_metrics[1, 1]), 2)
    accuracy_ = metrics.accuracy_score(y_test, y_pred)
    error_rate_ = 1 - accuracy_
    false_positive_rate_ = round(float(y_metrics[0,1]) / float(y_metrics[0,0]+y_metrics[0,1]), 2)
    F_score_ = (1+beta*beta)*(precision_1_ * recall_1_)/(beta*beta*precision_1_+recall_1_)
    type1_ = 1 - recall_1_
    type2_ = 1 - recall_0_
    precision_0.append(precision_0_)
    precision_1.append(precision_1_)
    recall_0.append(recall_0_)
    recall_1.append(recall_1_)
    accuracy.append(accuracy_)
    error_rate.append(error_rate_)
    false_positive_rate.append(false_positive_rate_)
    F_score.append(F_score_)
    type1.append(type1_)
    type2.append(type2_)


    clf = DecisionTreeClassifier(max_depth=4)
    clf = clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    y_metrics = metrics.confusion_matrix(y_test, y_pred)
    precision_0_ = round(float(y_metrics[0, 0]) / float(y_metrics[0, 0] + y_metrics[1, 0]), 2)
    precision_1_ = round(float(y_metrics[1, 1]) / float(y_metrics[0, 1] + y_metrics[1, 1]), 2)
    recall_0_ = round(float(y_metrics[0, 0]) / float(y_metrics[0, 0] + y_metrics[0, 1]), 2)
    recall_1_ = round(float(y_metrics[1, 1]) / float(y_metrics[1, 0] + y_metrics[1, 1]), 2)
    accuracy_ = metrics.accuracy_score(y_test, y_pred)
    error_rate_ = 1 - accuracy_
    false_positive_rate_ = round(float(y_metrics[0,1]) / float(y_metrics[0,0]+y_metrics[0,1]), 2)
    F_score_ = (1+beta*beta)*(precision_1_ * recall_1_)/(beta*beta*precision_1_+recall_1_)
    type1_ = 1 - recall_1_
    type2_ = 1 - recall_0_
    precision_0.append(precision_0_)
    precision_1.append(precision_1_)
    recall_0.append(recall_0_)
    recall_1.append(recall_1_)
    accuracy.append(accuracy_)
    error_rate.append(error_rate_)
    false_positive_rate.append(false_positive_rate_)
    F_score.append(F_score_)
    type1.append(type1_)
    type2.append(type2_)



    clf = GradientBoostingClassifier(max_depth=4)
    clf = clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    y_metrics = metrics.confusion_matrix(y_test, y_pred)
    precision_0_ = round(float(y_metrics[0, 0]) / float(y_metrics[0, 0] + y_metrics[1, 0]), 2)
    precision_1_ = round(float(y_metrics[1, 1]) / float(y_metrics[0, 1] + y_metrics[1, 1]), 2)
    recall_0_ = round(float(y_metrics[0, 0]) / float(y_metrics[0, 0] + y_metrics[0, 1]), 2)
    recall_1_ = round(float(y_metrics[1, 1]) / float(y_metrics[1, 0] + y_metrics[1, 1]), 2)
    accuracy_ = metrics.accuracy_score(y_test, y_pred)
    error_rate_ = 1 - accuracy_
    false_positive_rate_ = round(float(y_metrics[0,1]) / float(y_metrics[0,0]+y_metrics[0,1]), 2)
    F_score_ = (1+beta*beta)*(precision_1_ * recall_1_)/(beta*beta*precision_1_+recall_1_)
    type1_ = 1 - recall_1_
    type2_ = 1 - recall_0_
    precision_0.append(precision_0_)
    precision_1.append(precision_1_)
    recall_0.append(recall_0_)
    recall_1.append(recall_1_)
    accuracy.append(accuracy_)
    error_rate.append(error_rate_)
    false_positive_rate.append(false_positive_rate_)
    F_score.append(F_score_)
    type1.append(type1_)
    type2.append(type2_)



    clf = KNeighborsClassifier()
    X_train_scaled = preprocessing.scale(X_train)
    X_train_scaled = pd.DataFrame(X_train_scaled, columns=input.columns)
    X_test_scaled = preprocessing.scale(X_test)
    X_test_scaled = pd.DataFrame(X_test_scaled, columns=input.columns)
    clf = clf.fit(X_train_scaled, y_train)
    y_pred = clf.predict(X_test_scaled)
    y_metrics = metrics.confusion_matrix(y_test, y_pred)
    precision_0_ = round(float(y_metrics[0, 0]) / float(y_metrics[0, 0] + y_metrics[1, 0]), 2)
    precision_1_ = round(float(y_metrics[1, 1]) / float(y_metrics[0, 1] + y_metrics[1, 1]), 2)
    recall_0_ = round(float(y_metrics[0, 0]) / float(y_metrics[0, 0] + y_metrics[0, 1]), 2)
    recall_1_ = round(float(y_metrics[1, 1]) / float(y_metrics[1, 0] + y_metrics[1, 1]), 2)
    accuracy_ = metrics.accuracy_score(y_test, y_pred)
    error_rate_ = 1 - accuracy_
    false_positive_rate_ = round(float(y_metrics[0,1]) / float(y_metrics[0,0]+y_metrics[0,1]), 2)
    F_score_ = (1+beta*beta)*(precision_1_ * recall_1_)/(beta*beta*precision_1_+recall_1_)
    type1_ = 1 - recall_1_
    type2_ = 1 - recall_0_
    precision_0.append(precision_0_)
    precision_1.append(precision_1_)
    recall_0.append(recall_0_)
    recall_1.append(recall_1_)
    accuracy.append(accuracy_)
    error_rate.append(error_rate_)
    false_positive_rate.append(false_positive_rate_)
    F_score.append(F_score_)
    type1.append(type1_)
    type2.append(type2_)



    clf = LogisticRegression()
    X_train_scaled = preprocessing.scale(X_train)
    X_train_scaled = pd.DataFrame(X_train_scaled, columns=input.columns)
    X_test_scaled = preprocessing.scale(X_test)
    X_test_scaled = pd.DataFrame(X_test_scaled, columns=input.columns)
    clf = clf.fit(X_train_scaled, y_train)
    y_pred = clf.predict(X_test_scaled)
    y_metrics = metrics.confusion_matrix(y_test, y_pred)
    precision_0_ = round(float(y_metrics[0, 0]) / float(y_metrics[0, 0] + y_metrics[1, 0]), 2)
    precision_1_ = round(float(y_metrics[1, 1]) / float(y_metrics[0, 1] + y_metrics[1, 1]), 2)
    recall_0_ = round(float(y_metrics[0, 0]) / float(y_metrics[0, 0] + y_metrics[0, 1]), 2)
    recall_1_ = round(float(y_metrics[1, 1]) / float(y_metrics[1, 0] + y_metrics[1, 1]), 2)
    accuracy_ = metrics.accuracy_score(y_test, y_pred)
    error_rate_ = 1 - accuracy_
    false_positive_rate_ = round(float(y_metrics[0,1]) / float(y_metrics[0,0]+y_metrics[0,1]), 2)
    F_score_ = (1+beta*beta)*(precision_1_ * recall_1_)/(beta*beta*precision_1_+recall_1_)
    type1_ = 1 - recall_1_
    type2_ = 1 - recall_0_
    precision_0.append(precision_0_)
    precision_1.append(precision_1_)
    recall_0.append(recall_0_)
    recall_1.append(recall_1_)
    accuracy.append(accuracy_)
    error_rate.append(error_rate_)
    false_positive_rate.append(false_positive_rate_)
    F_score.append(F_score_)
    type1.append(type1_)
    type2.append(type2_)



    clf = MLPClassifier(solver='lbgfs')
    X_train_scaled = preprocessing.scale(X_train)
    X_train_scaled = pd.DataFrame(X_train_scaled, columns=input.columns)
    X_test_scaled = preprocessing.scale(X_test)
    X_test_scaled = pd.DataFrame(X_test_scaled, columns=input.columns)
    clf = clf.fit(X_train_scaled, y_train)
    y_pred = clf.predict(X_test_scaled)
    y_metrics = metrics.confusion_matrix(y_test, y_pred)
    precision_0_ = round(float(y_metrics[0, 0]) / float(y_metrics[0, 0] + y_metrics[1, 0]), 2)
    precision_1_ = round(float(y_metrics[1, 1]) / float(y_metrics[0, 1] + y_metrics[1, 1]), 2)
    recall_0_ = round(float(y_metrics[0, 0]) / float(y_metrics[0, 0] + y_metrics[0, 1]), 2)
    recall_1_ = round(float(y_metrics[1, 1]) / float(y_metrics[1, 0] + y_metrics[1, 1]), 2)
    accuracy_ = metrics.accuracy_score(y_test, y_pred)
    error_rate_ = 1 - accuracy_
    false_positive_rate_ = round(float(y_metrics[0,1]) / float(y_metrics[0,0]+y_metrics[0,1]), 2)
    F_score_ = (1+beta*beta)*(precision_1_ * recall_1_)/(beta*beta*precision_1_+recall_1_)
    type1_ = 1 - recall_1_
    type2_ = 1 - recall_0_
    precision_0.append(precision_0_)
    precision_1.append(precision_1_)
    recall_0.append(recall_0_)
    recall_1.append(recall_1_)
    accuracy.append(accuracy_)
    error_rate.append(error_rate_)
    false_positive_rate.append(false_positive_rate_)
    F_score.append(F_score_)
    type1.append(type1_)
    type2.append(type2_)



    clf = RandomForestClassifier()
    clf = clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    y_metrics = metrics.confusion_matrix(y_test, y_pred)
    precision_0_ = round(float(y_metrics[0, 0]) / float(y_metrics[0, 0] + y_metrics[1, 0]), 2)
    precision_1_ = round(float(y_metrics[1, 1]) / float(y_metrics[0, 1] + y_metrics[1, 1]), 2)
    recall_0_ = round(float(y_metrics[0, 0]) / float(y_metrics[0, 0] + y_metrics[0, 1]), 2)
    recall_1_ = round(float(y_metrics[1, 1]) / float(y_metrics[1, 0] + y_metrics[1, 1]), 2)
    accuracy_ = metrics.accuracy_score(y_test, y_pred)
    error_rate_ = 1 - accuracy_
    false_positive_rate_ = round(float(y_metrics[0,1]) / float(y_metrics[0,0]+y_metrics[0,1]), 2)
    F_score_ = (1+beta*beta)*(precision_1_ * recall_1_)/(beta*beta*precision_1_+recall_1_)
    type1_ = 1 - recall_1_
    type2_ = 1 - recall_0_
    precision_0.append(precision_0_)
    precision_1.append(precision_1_)
    recall_0.append(recall_0_)
    recall_1.append(recall_1_)
    accuracy.append(accuracy_)
    error_rate.append(error_rate_)
    false_positive_rate.append(false_positive_rate_)
    F_score.append(F_score_)
    type1.append(type1_)
    type2.append(type2_)



    clf = SVC()
    X_train_scaled = preprocessing.scale(X_train)
    X_train_scaled = pd.DataFrame(X_train_scaled, columns=input.columns)
    X_test_scaled = preprocessing.scale(X_test)
    X_test_scaled = pd.DataFrame(X_test_scaled, columns=input.columns)
    clf = clf.fit(X_train_scaled, y_train)
    y_pred = clf.predict(X_test_scaled)
    y_metrics = metrics.confusion_matrix(y_test, y_pred)
    precision_0_ = round(float(y_metrics[0, 0]) / float(y_metrics[0, 0] + y_metrics[1, 0]), 2)
    precision_1_ = round(float(y_metrics[1, 1]) / float(y_metrics[0, 1] + y_metrics[1, 1]), 2)
    recall_0_ = round(float(y_metrics[0, 0]) / float(y_metrics[0, 0] + y_metrics[0, 1]), 2)
    recall_1_ = round(float(y_metrics[1, 1]) / float(y_metrics[1, 0] + y_metrics[1, 1]), 2)
    accuracy_ = metrics.accuracy_score(y_test, y_pred)
    error_rate_ = 1 - accuracy_
    false_positive_rate_ = round(float(y_metrics[0,1]) / float(y_metrics[0,0]+y_metrics[0,1]), 2)
    F_score_ = (1+beta*beta)*(precision_1_ * recall_1_)/(beta*beta*precision_1_ + recall_1_)
    type1_ = 1 - recall_1_
    type2_ = 1 - recall_0_
    precision_0.append(precision_0_)
    precision_1.append(precision_1_)
    recall_0.append(recall_0_)
    recall_1.append(recall_1_)
    accuracy.append(accuracy_)
    error_rate.append(error_rate_)
    false_positive_rate.append(false_positive_rate_)
    F_score.append(F_score_)
    type1.append(type1_)
    type2.append(type2_)


    result = [precision_0, precision_1, recall_0, recall_1, accuracy, error_rate, false_positive_rate, F_score, type1, type2]
    result = pd.DataFrame(result, columns=method, index=['precision_0', 'precision_1', 'recall_0', 'recall_1',
                                                         'accuracy', 'error_rate', 'false_positive_rate', 'F_score', 'type1', 'type2'])
    result = round(result, 2)
    print(result)
```
